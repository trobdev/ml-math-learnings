## Review Vectors and Dot Products

Before we move ahead, let's recap on some of the most essential knowledge from the Part I notebook, Vectors and Similarity. 

- What is a vector?  A vector is a way of measuring displacement using distance and direction.  Vectors exist within dimensional space, expressed as $\textbf{v}, \textbf{w} \in \mathbb{R{n}}$.
- What does a dot product compute geometrically?  From a geometric perspective it computes the amount of overlap between vector $v$ and vector $w$, and then produces a scalar from it.  The further $w$ moves from $v$, the closer the dot product gets to 0.0.
- Why does a dot product show up in a linear model?  It's an effective means of measuring similarity within any dimensional space, but it's very easily expressed and understood within 2D space (linear). We'll dive into this further from an ML perspective...
- What is the difference between a vector's **norm** and a **normalized** vector?  A vector's **norm** is simply a measurement of a vector's length in space.  It's useful because you can effectively measure it in any dimension.  A **normalized** vector is a way of comparing vectors where distance is less important than direction.  Normalizing a vector means reducing its length to 1, while maintaining direction.  This is useful for cosine similarity.

```python
# Recap of dot product and norm
vec1 = [1, 2]
vec2 = [3, 4]

def dot_product(v, w):
    if len(v) != len(w):
        raise ValueError('Vectors are not of equal length')
    product = 0
    for i in range(len(v)):
        product += v[i] * w[i]
    return product

def norm(v):
    return (dot_product(v, v))**0.5

print('Dot Product:', dot_product(vec1, vec2)) #Expect 11
print('Norm of V:', norm(vec1)) #Expect 2.24
```

## Matrices as an Organized Collection of Vectors

Matrices provide a way to store datasets in the form of a 2D rectangular array of numbers, where each **row** is an example and each **column** is a feature.  

In the seminal work 'Deep Learning' by Goodfellow, Bengio, and Courville the convention for matrices is described below:  

If $\textbf{A} \in \mathbb{R^{m \times n}}$, then:
- $\textbf{A}$ = matrix (uppercase, vs. lowercase for vector)
- $m$ = number of examples (rows)
- $n$ = number of features (columns)

$\textbf{A}_{1, 1}$ would define the upper left entry, where $\textbf{A}_{m, n}$ would define the lower right entry. It is also common to see $i$ and $j$ used as the corresponding equivalent to row, column. (*"Deep Learning." Goodfellow, Bengio, Courville. p. 30-1*) 

Matrix vs. vector:


```python
# Example matrix
A = [[1.0, 2.0],
     [2.0, 1.0],
     [3.0, 4.0]]
x = [0.5, 1.5]
print('A =', A)
print('x =', x)
```

A simple rule of matrices is that each row must have the same number of values as each other row, and the same goes for columns.  If a matrix is a 3 x 3 matrix, then there should be 9 values present for it to be legitimate.  The $m \times n$ is referred to as the "size" of the matrix.  A real matrix can be expressed as $\mathcal{M}$ where $m$ and $n$ likewise represent rows and columns respectively (*"Introduction to Linear and Matrix Algebra." Johnston. p. 21*) 

### Shape

Knowing that each row has the same number of columns, we can easily figure out what the "shape" of a matrix is, as well as a vector.  Understanding the shape of a matrix is important because it is a constraint unto itself.  Much like vectors, multiplication of matrices has its own limitations that we'll get into later.  For now let's see how we can identify shape in Python:

```python
# Shape of Matrix
def shape_matrix(M):
    if len(M) == 0:
        return (0, 0)
    return (len(M), len(M[0]))

def shape_vector(v):
    return (len(v))

print('Shape of A:', shape_matrix(A)) # Expect 3,2
print('Shape of x:', shape_vector(x)) # Expect 2 (1 row, 2 columns, 2 unique entries)
```

For matrix $A$ we could say that it has 3 examples and 2 features per example. This has direct applicability to Machine Learning, which relies on features to draw patterns and find relationships between pieces of raw data.   

### Matrix Addition and Scalar Multiplication

Matrix addition and scalar multiplication are very similar to what we encountered already with vectors. 

In order to be thorough, here's a quick example of how these would be expressed:

Addition: $[{A} + {B}]_{i, j} = a_{i,j} + b_{i, j}$

$$
\begin{bmatrix} 
1 & 2 \\
2 & 3
\end{bmatrix}
+ 
\begin{bmatrix}
2 & 3 \\
3 & 4
\end{bmatrix}
=
\begin{bmatrix}
3 & 5 \\
5 & 7
\end{bmatrix}
$$

Scalar Multiplication: $[c{A}]_{i,j} = ca_{i, j}$

$$
2A = 2 \cdot \begin{bmatrix} 1 & 2 \\ 3 & 4 \end{bmatrix} = \begin{bmatrix} 2 & 4 \\ 6 & 8 \end{bmatrix}
$$


## The Transpose

A **transpose** swaps rows and columns with one another.  It effectively creates a mirror image of a matrix across its left to right diagonal line.  What is important is that while the shape is changed, the contents are not.  This is useful in matrix multiplication because it's a component of multiplying one matrix by another.  

A real matrix $\mathcal{M_{m, n}}$ would become $\mathcal{M_{n, m}}$. 

We can express a transpose like so: $A^{T}$

The transpose is used all the time in Linear Algebra and therefore ML as well.  We'll encounter it over and over again under the hood, but for now it's best to understand what it does in code vs. all of its implications:

```python
# Calculate the transpose

def transpose(M):
    rows, cols = shape_matrix(M) # Leverage existing shape function
    return [[M[r][c] for r in range(rows)] for c in range(cols)]

print('A =', A)
print('A^T =', transpose(A)) # Expect [[1.0, 2.0, 3.0], [2.0, 1.0, 4.0]]
```

## Matrix Multiplication

Matrix multiplication is still multiplication, but the method differs from vector multiplication considerably.  Perhaps the most notable complication to it is that you can only multiply matrix product $AB$ when $A$ has the same number of columns as $B$ has rows.  

From Johnston's definition of Matrix Multiplication (*Ibid., p. 21*): 

Assuming $A$ and $B$ are both real matrices, the product $AB$ is $m \times p$ matrix whose $(i, j)$ entry for each $1 \le i \le m$ and $1 \le j \le p$, is: 
$$[AB]_{i,j} = a_{i, 1}b_{1, j} + a_{i, 2}b_{2, j} + ... + a_{i, n}b_{n, j}$$

To calculate this, we determine dot products of the Rows of $A$ with the Columns of $B$.  As we said earlier, it only works when numColumnsA = numRowsB.  

Let's write it out in Python:  

```python
# Define A and B matrices
matA = [[1, 3],
        [5, 7]]
matB = [[2, 4],
        [6, 8]]

# Perform matrix-matrix multiplication
def matmul(A, B):
    rows_A, cols_A = len(A), len(A[0])
    rows_B, cols_B = len(B), len(B[0])

    if cols_A != rows_B:
        raise ValueError(f'Matrices are incompatible.  A is {rows_A} x {cols_A}, B is {rows_B} x {cols_B}')
    
    B_T = transpose(B) # Leverage existing transpose function
    new_mat = []
    for row in A:
        new_row = []
        for col in B_T:
            new_row.append(dot_product(row, col)) # Leverage existing dot product function
        new_mat.append(new_row)
    return new_mat
    
print(matmul(matA, matB)) # Expect [[20, 28], [52, 76]]
    
```

As we can see, matrix multiplication builds on all of the previous functions we used here to product a result: `dot_product()`, `shape()`, and `transpose()`. 

There are of course other ways to write this function as a standalone, but together it's possible to perform that multiplication using what we've made so far.  

### Matrix Vector Multiplication 

It is also possible to multiply matrices by vectors.  It bears similiarity with matrices, but with a few less steps involved due to much lower complexity.  One reason you might do that is to simply move the vector elsewhere, because doing so alters the basis vector (starting point).  

Shape is still important, in that if the length of a vector $x$ does not equal the number of columns of a matrix $A$, then we cannot multiply them together.  

We can express this like so:
$$
Av = b
$$

We'll write it quickly in Python:

```python
# Define vector
vecX = [3, 3]

# Matrix Vector multiplication
def matvec(A, x):
    rows, cols = len(A), len(A[0])
    if len(x) != cols:
        raise ValueError(f'Shapes are incompatible.  A is {rows} x {cols}, x has length {len(x)}')

    new_vec = []
    for row in A:
        new_vec.append(dot_product(row, x))
    return new_vec

print("Ax =", matvec(matA, vecX)) # Expect [12, 36]
```

Matrix vector multiplication can produce two different categories of matrices:  The **column vector** and the **row vector** (*Ibid., p. 27*).  Much like they sound, they look like the following: 

$$
\begin{aligned}
v = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} \\
\\
w = \begin{bmatrix} 1 & 2 & 3 \end{bmatrix}
\end{aligned}
$$

These behave very differently and have interesting utility as inputs and tests that we'll explore later. 

## Creating a Linear Model

Now we're going to make a linear model, which is a simple predictive model that has a lot of applicability within ML (which is about predicting outcomes based on data).  It also serves as a way to tie together all the vector and matrix concepts we've explored so far.  

### Why a Linear Model?

Linear models are helpful for both **predicting** (including *classification* and *regression*), as well as **understanding** how features influence target (dependent) variables.  

Most of us are exposed to the classic $y = mx + b$ to express a linear relationship of a simple line drawn on a 2D graph.  A linear model here is just that, only we're taking that and changing the verbiage to better suit ML use cases. 

Instead of the classic formula, we express it as so: $\hat{y} = w^{T}x+b$

Where:

- $\hat{y}$ = the model's prediction
- $w$ = weight vector (feature importance)
- $x$ = feature vectors
- $b$ = bias/intercept

Our prediction then depends on the weights we assign to features, and the biases inherent.  Some features will have less weight, some biases might be greater or lesser.  But nonetheless we walk away with something resembling a very common equation.

If we're dealing with many examples (many rows) at once, then we can express it slightly differently:$\hat{y} = Xw + b$

Where:

- $X \in \mathbb{R^{nxd}}$: $n$ = examples/rows, $d$ = features/columns
- $w \in \mathbb{R^{d}}$: weights
- $b \in \mathbb{R}$: bias

So that one matrix-vector multiplication $Xw$ will compute all dot products $w^Tx_i$ at once.  It's a matter of effiency because using matrix form instead allows us to express it in a more compact, repeatable way. 

We can use linear models for so many things in ML:  predicting prices, predicting heart disease, classifying images, etc.  Many of the `scikit-learn` algorithms are linear models!

Now let's code it in Python:

```python
# Simple linear prediction
def predict_linear(X, w, b=0.0):
    linear_component = matvec(X, w)
    return [value + b for value in linear_component]

X = [[1.0],
     [2.0],
     [3.0]]

w = [2.0]
b = 1.0

y_preds = predict_linear(X, w, b)
print("Predictions =", y_preds) # Expect [3.0, 5.0, 7.0]
```

## Apply the Linear Model to a Toy Dataset

To see this in action and gain a better understanding of what it is really doing and how it is responding, we'll create a toy dataset to predict racing times for a 1 mile / 1600M race.  

As a former runner, we greatly enjoyed carbo-loading pizza and sleeping a lot before races, so with that in mind let's put something together where our features are [slices_pizzas_ate, hours_slept] as $X_{i,j}$ and time in minutes for $\hat{y}$.   

```python
# Toy dataset based on predicting running times
# [ slices_pizzas ate, hours_slept ]
X = [[4.0, 6.0],
     [3.0, 8.0],
     [5.0, 7.0],
     [3.0, 6.0],
     [2.0, 7.0],
     [3.0, 5.0]]

# [ time_in_minutes ]
y_true = [5.5, 5.0, 5.8, 5.5, 4.9, 5.7] # Matches numRows

# Guessed model
w = [1.0, -1.0] # Estimating that for 1 slice of pizza eaten, y would increase by 1 minute; conversely, sleeping an extra hour would reduce time by 1 minute
b = 10 # Estimating that if x = 0, it would take me 10 minutes to run that mile (no sleep means awful results, trust me)

y_pred = predict_linear(X, w, b)

for x_i, y_t, y_p in zip(X, y_true, y_pred):
    print(f'X = {x_i}, true = {y_t}, pred = {y_p}, residual = {y_t - y_p}')
```

Although simple, we're seeing some interesting results here.  The model tends to overshoot by a bit, but does demonstrate that in general more pizza = slower predicted time, and more sleep = faster predicted time.  Where it seems to have the most issues is when sleep is low relative to number of pizza slices.  There's a relationship there that if further developed by more data might iron itself out.  

We can take and apply this process to much, much larger datasets but probably not like this.  We can stack multiple of these linear predictors together where we have not just an input matrix, but also a weight matrix.  If we were looking for multiple linear scores/multiple outputs for each example, that would be the preferred way.  Neural network layers rely on this.  

## Bonus Section: Scikit-Learn vs. Custom Model

While I said earlier I won't be using any custom libraries, I am going to make a slight tweak here because I think it's worth showing for demonstration purposes only.  We've done the learning, we can certainly afford a comparison at this point.

`Scikit-learn` is a popular classical ML library that has many different models suitable to supervised learning tasks (at least that's been my exposure to it).  Usually it's a first stop for getting your hands dirty with ML. We are introducing here to show what it can do vs. what we just did. 

Let's apply the same concepts using SKL and do a little comparison:

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Toy data and manual parameters
X_skl = np.array(X, dtype=float)      # shape (n_samples, n_features)
y_skl = np.array(y_true, dtype=float) # defined only for reference since we set coef_ and intercept_ manually

w_manual = np.array(w, dtype=float)  # Our chosen weights, length n_features
b_manual = float(b)                  # Our chosen bias

# 2. Our prediction from earlier, updated to use numpy instead
y_hat_custom = np.array(predict_linear(X, w_manual.tolist(), b_manual))

# 3. Builds an sklearn model with the SAME w and b as earlier
sk_model = LinearRegression(fit_intercept=True)  # we set intercept manually instead of using the standard model.fit() function
sk_model.coef_ = w_manual.reshape(1, -1)          # shape (1, n_features)
sk_model.intercept_ = b_manual                    # setting intercept to our earlier one (10)

y_hat_sklearn = sk_model.predict(X_skl).ravel() # Flattens our array

print("Custom preds :", y_hat_custom)
print("Sklearn preds:", y_hat_sklearn)
```

As we can see, we're now getting the same results using the scikit-learn `LinearRegression()` model with the same weights and biases.  You can read more about it here: https://scikit-learn.org/stable/modules/linear_model.html.  Like we learned earlier, linear models are prevalent and used under the hood of many popular libraries.  They might differ in sophistication, but they have broad utility.
